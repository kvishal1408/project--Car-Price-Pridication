# Model Building

```python
x=data_1[['Transmission','KM','Owner','Fuel_type','Age','Year','EMI','Brand','Name','Model','Registration']]
y=data_1['Price']
```

# Testing Model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['Transmission','KM','Owner','Fuel_type','Age','Year','EMI','Brand','Name','Model','Registration']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Transmission','KM','Owner','Fuel_type','Age','Year','EMI','Brand','Name','Model','Registration']),
                                    remainder='passthrough')
```

# Linear Regression Model

```python
lr=LinearRegression()
```

```python
Model_Lr=make_pipeline(column_trans,lr)
```

# Fitting the model

```python
Model_Lr.fit(x_train,y_train)
```

```python
y_pred=Model_Lr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    Model_Lr=make_pipeline(column_trans,lr)
    Model_Lr.fit(x_train,y_train)
    y_pred=Model_Lr.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
Model_Lr=make_pipeline(column_trans,lr)
Model_Lr.fit(x_train,y_train)
y_pred=Model_Lr.predict(x_test)
r2_score(y_test,y_pred)*100
```
